```python
class TicTacToe:

    def __init__(self):
        pass

    # Printing the Board of TicTacToe
    def printBoard(self):        
        print(" {} | {} | {}".format(self.outputArray[0],self.outputArray[1],self.outputArray[2]))
        print(u'\u2015\u2015\u2015 \u2015\u2015\u2015 \u2015\u2015\u2015')
        print(" {} | {} | {}".format(self.outputArray[3],self.outputArray[4],self.outputArray[5]))
        print(u'\u2015\u2015\u2015 \u2015\u2015\u2015 \u2015\u2015\u2015')
        print(" {} | {} | {}".format(self.outputArray[6],self.outputArray[7],self.outputArray[8]))
    
    # Logic for TicTacToe Game
    def inputArray(self):
        Player1 = "x"
        Player2 = "O"
        self.outputArray = [" "," "," "," "," "," "," "," "," "]
        isFirstPlayer = True  
        isFull = False
        isWon = False
        while(isFull == False):
            if(self.isCheckFull() == True):
                isFull = True
                print("Game Draw!")
                break
            if(isFirstPlayer == True):
                position = int(input("What position player1 want to enter:"))
                if(self.outputArray[position-1] == "X" or self.outputArray[position-1] == "O"):
                    posOccupied = self.outputArray[position-1]
                    print("Position Already Occuipied. Enter another position!")
                else:
                    self.outputArray[position-1] = "X"
                    self.printBoard()
                    isWon = self.winningChance("X")
                    isFirstPlayer = False
                    if(isWon == True):
                        print("Player1 Won!")
                        break
            
            elif(isFirstPlayer == False):
                position = int(input("What position Player2 want to enter:"))
                if(self.outputArray[position-1] == "X" or self.outputArray[position-1] == "O"):
                    posOccupied = self.outputArray[position-1]
                    print("Position Already Occuipied. Enter another position!")
                else:
                    self.outputArray[position-1] = "O"
                    self.printBoard()
                    isWon = self.winningChance("O")
                    isFirstPlayer = True
                    if(isWon == True):
                        print("Player2 Won!")
                        break
    
    #To check the board is full
    def isCheckFull(self):        
        x = len(self.outputArray)
        for i in range(x):
            if(self.outputArray[i] == " "):
                return False
        return True
    
    # Winning Chances of the Players
    def winningChance(self, player):
        if(self.outputArray[0] == player and self.outputArray[1] == player and self.outputArray[2] == player):
            return True
        elif(self.outputArray[0] == player and self.outputArray[3] == player and self.outputArray[6] == player):
            return True
        elif(self.outputArray[0] == player and self.outputArray[4] == player and self.outputArray[8] == player):
            return True
        elif(self.outputArray[1] == player and self.outputArray[4] == player and self.outputArray[7] == player):
            return True
        elif(self.outputArray[2] == player and self.outputArray[5] == player and self.outputArray[8] == player):
            return True
        elif(self.outputArray[3] == player and self.outputArray[4] == player and self.outputArray[5] == player):
            return True
        elif(self.outputArray[6] == player and self.outputArray[7] == player and self.outputArray[8] == player):
            return True
        elif(self.outputArray[2] == player and self.outputArray[4] == player and self.outputArray[6] == player):
            return True
        
```
